 # PCA vs t-SNE

What are PCA & t-SNE?

PCA: Linear dimensionality reduction technique that identifies the axes (principal components) maximizing the variance of the data. 

- A statistical procedure that employs orthogonal transformation to convert a set of observations of correlated variables into a set of values of linearly uncorrelated variables called principal components.
- It's widely used for dimensionality reduction without losing much of the data's variability.

t-SNE: Non-linear dimensionality reduction technique used to visualize high-dimensional data by mapping similar points close together.
- A machine learning algorithm for visualization.
- It is particularly well-suited for visualization of high-dimensional datasets.

# Setup

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn import datasets
```

# Sample Data

```python
digits = datasets.load_digits()
```

```python
X = digits.data
y = digits.target
```

```python
X
```

```python
y
```

# PCA

We're applying PCA to reduce the dataset from 64 dimensions to just 2 dimensions, making it easier to visualize.

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='jet', edgecolor='k', s=40)
plt.colorbar()
plt.title("PCA Visualization")
plt.show()
```

# TSNE


Here, we use t-SNE to reduce the dataset to 2 dimensions.

t-SNE's parameters, like perplexity and n_iter, play a vital role in the visualization. Adjusting these can lead to different results.

- perplexity: Relates to the number of nearest neighbors t-SNE considers when generating the visualization.
- n_iter: Number of iterations for the optimization.

```python
tsne = TSNE(n_components=2, random_state=0, perplexity=30, n_iter=300)
X_tsne = tsne.fit_transform(X)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='jet', edgecolor='k', s=40)
plt.colorbar()
plt.title("t-SNE Visualization")
plt.show()
```
